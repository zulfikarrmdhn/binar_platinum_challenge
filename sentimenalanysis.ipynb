```python
df = pd.read_csv('/content/train_preprocess.tsv',sep='\t',names=['Komentar', 'Kategori'], encoding='latin1')
```

```python
df_clean_duplicates = df.drop_duplicates()
```

```python
df_clean_tb = df_clean_duplicates.copy()
```

```python
df_clean_tb
```

```python

df_clean_tb['Komentar'] = df_clean_tb['Komentar'].str.strip()
df_clean_tb['Komentar'] = df_clean_tb['Komentar'].replace(r'([A-Z]+)\s(\d+)', r'\1\2', regex=True)
df_clean_tb['Komentar'] = df_clean_tb['Komentar'].str.replace(r'\\x[A-Za-z0-9./:)(*&^%$#@!_;]+', '')
df_clean_tb['Komentar'] = df_clean_tb['Komentar'].str.replace(r'[^\w\d\s]+', '')
df_clean_tb['Komentar'] = df_clean_tb['Komentar'].str.lower()
df_clean_tb['Komentar'] = df_clean_tb['Komentar'].str.replace(r'user', '')
df_clean_tb['Komentar'] = df_clean_tb['Komentar'].str.replace(r'_', '')
```
